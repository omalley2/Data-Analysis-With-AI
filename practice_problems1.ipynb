## 11. Mini-Project 3: Quiz Function
Write a function that takes the below questions
1. Asks the users the question with the possible answers
2. Tells the user whether they got the question right or now
3. Calculates an overall score

```python
# Define quiz questions
quiz = [
    {"question": "What is the capital of France?", "options": ["Paris", "London", "Berlin", "Rome"], "answer": "Paris"},
    {"question": "Which language is used for data science?", "options": ["Python", "Java", "C++", "Ruby"], "answer": "Python"},
    {"question": "What does 'int' stand for?", "options": ["Integer", "Interval", "Internal", "Internet"], "answer": "Integer"}
]

# Function to run the quiz
def run_quiz(questions):
    score = 0
    for q in questions:
        print("\n" + q["question"])
        for i, option in enumerate(q["options"], start=1):
            print(f"{i}. {option}")
        choice = input("Enter the number of your answer: ")
        if q["options"][int(choice)-1] == q["answer"]:
            print("Correct!")
            score += 1
        else:
            print(f"Incorrect. The correct answer was: {q['answer']}")
    print(f"\nYour final score: {score}/{len(questions)}")

# Run the quiz
run_quiz(quiz)
```

## 12.1

1. Write a script that reads a list of numbers from the user, then:
   - Prints only the prime numbers (use a helper function)  
   - Uses a `for` loop and a list comprehension

```python
def is_prime(n):
    """
    Return True if n is prime, False otherwise.
    """
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Read input
nums = input("Enter numbers separated by spaces: ").split()
nums = [int(x) for x in nums]

# 1. Using a for‐loop
primes_loop = []
for num in nums:
    if is_prime(num):
        primes_loop.append(num)
print("Primes (for loop):", primes_loop)

# 2. Using a list comprehension
primes_lc = [num for num in nums if is_prime(num)]
print("Primes (list comprehension):", primes_lc)
```

## 12.2

2. Build a simple contact manager:
   - Define functions to add, list, and search contacts (stored in a list of dicts)  
   - Use control flow and loops for menu navigation

```python
contacts = []

def add_contact(name, email, phone):
    contacts.append({"name": name, "email": email, "phone": phone})
    print(f"Added contact: {name}")

def list_contacts():
    if not contacts:
        print("No contacts to display.")
    else:
        print("\nAll Contacts:")
        for i, c in enumerate(contacts, 1):
            print(f"{i}. {c['name']} – {c['email']} – {c['phone']}")
    print()

def search_contacts(query):
    results = [c for c in contacts if query.lower() in c["name"].lower()]
    if results:
        print("\nSearch Results:")
        for c in results:
            print(f"{c['name']} – {c['email']} – {c['phone']}")
    else:
        print("No matching contacts found.")
    print()

def menu():
    while True:
        print("Contact Manager Menu:")
        print("1. Add contact")
        print("2. List contacts")
        print("3. Search contacts")
        print("4. Exit")
        choice = input("Choose an option: ")
        if choice == "1":
            name = input("Name: ")
            email = input("Email: ")
            phone = input("Phone: ")
            add_contact(name, email, phone)
        elif choice == "2":
            list_contacts()
        elif choice == "3":
            query = input("Enter name to search: ")
            search_contacts(query)
        elif choice == "4":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Try again.\n")


menu()
```
